SCATTER PLOT

one of the graphs used to represent a relationship

```python
import matplotlib.pyplot as plt
```

```python
# alpha reduces the size of the dot
plt.plot( 'height', 'weight', 'o', alpha=0.02)
```

```python
# plotting a scatter diagram
df=pd.read_excel(r"studentdata.xlsx")
h = df['height']
w =df['weight']
plt.plot( h,w, 'o')
```

```python
plt.title("Relationship between weight and height in Cohort 3")
plt.xlabel("weight in kg")
plt.ylabel("height in cm")
```

```python
df.height = 'height'
df.weight = 'weight'
plt.plot('height', 'weight' )
```
